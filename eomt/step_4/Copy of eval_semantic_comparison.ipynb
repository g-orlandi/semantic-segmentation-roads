```python
import torch
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")

PROJECT_DIR = "/content/drive/MyDrive/project/semantic-segmentation-roads/eomt"
state_dict_path_cs = "/content/drive/MyDrive/project/eomt_cityscapes.bin"
state_dict_path_coco = "/content/drive/MyDrive/project/eomt_coco.bin"
```

```python
import os, sys, importlib

os.chdir(PROJECT_DIR)
sys.path.insert(0, PROJECT_DIR)
importlib.invalidate_caches()

from my_eval import *
```

## Modello A (Cityscapes)

```python
config_cs = get_config()
data = get_data(config_cs)
eval_dataset_cs, img_size_cs, num_classes_cs = data.val_dataloader(), data.img_size, data.num_classes
model_cs = get_model(config_cs, img_size_cs, num_classes_cs).eval().to(device)
model_cs = load_weights(model_cs, state_dict_path_cs, device)
```

```python
evaluate_semantic(model_cs, eval_dataset_cs, device)
```

## Modello (COCO)

```python
config_coco = get_config(True)
data = get_data(config_coco)
eval_dataset_coco, img_size_coco, num_classes_coco = data.val_dataloader(), data.img_size, data.num_classes
model_coco = get_model(config_coco, img_size_coco, num_classes_coco, True).eval().to(device)
model_coco = load_weights(model_coco, state_dict_path_coco, device)
```
